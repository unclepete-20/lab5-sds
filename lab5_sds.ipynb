#### Parte 1 – Filtrado y preprocesamiento

```python
# Paso 1: Cargar librerías y archivos
import pandas as pd
from pandas.io.json import json_normalize
import json
```

```python
# Paso 2: Cargar la información del archivo large_even.json en una lista
with open('large_eve.json', 'r') as f:
    data = [json.loads(line) for line in f]

# Mostrar cantidad total de registros
total_registros = len(data)
print("Cantidad total de registros:", total_registros)
```

```python
# Paso 3: Filtrar únicamente registros DNS
dns_data = [record for record in data if record['event_type'] == 'dns' and 'dns' in record]

# Convertir la clave 'dns' en una lista de un solo elemento
for record in dns_data:
    record['dns'] = [record['dns']]

# Mostrar cantidad de registros DNS filtrados
registros_dns_filtrados = len(dns_data)
print("Cantidad de registros DNS filtrados:", registros_dns_filtrados)
```

```python
# Paso 4: Mostrar información de 2 registros cualesquiera
print("Información de 2 registros DNS:")
for i in range(2):
    print(json.dumps(dns_data[i], indent=4))
```

```python
# Paso 4: Normalizar la información y asignarla en un dataframe
dns_df = json_normalize(dns_data, record_path=['dns'], meta=['timestamp', 'flow_id', 'pcap_cnt', 'vlan', 'src_ip', 'src_port', 'dest_ip', 'dest_port', 'proto', 'dns'])

# Mostrar shape del dataframe
print("Shape del dataframe:", dns_df.shape)
```

```python
dns_df
```

```python
# Paso 6: Filtrar registros DNS tipo A
dns_df_a = dns_df[dns_df['rrtype'] == 'A']

# Mostrar cantidad de registros DNS tipo A
registros_dns_tipo_a = len(dns_df_a)
print("Cantidad de registros DNS tipo A:", registros_dns_tipo_a)
```

```python
# Paso 7: Filtrar dominios únicos
dominios_unicos = dns_df_a['rrname'].unique()
print("Cantidad de dominios únicos:", len(dominios_unicos))
```

Prompt de ChatGPT para obtener la funcion TLD:

"Podrias proporcionarme una función en Python que extraiga el TLD de un dominio?. Por ejemplo, al pasar "api.wunderground.com" como parámetro, la función debería retornar ".com"; y al pasar "safebrowsing.clients.google.com.home", debería retornar ".home"."

```python
# Paso 8: Definir función para obtener el TLD de un dominio
def obtener_TLD(dominio):
    partes = dominio.split('.')
    if len(partes) <= 1:
        return dominio
    else:
        return partes[-1]

# Ejemplo de uso de la función con ChatGPT
print("TLD para 'api.wunderground.com':", obtener_TLD("api.wunderground.com"))
print("TLD para 'safebrowsing.clients.google.com.home':", obtener_TLD("safebrowsing.clients.google.com.home"))
```

```python
# Paso 9: Obtener el TLD para cada dominio y crear nueva columna 'domain_tld'
dns_df_a_copy = dns_df_a.copy()  # Crear una copia del DataFrame
dns_df_a_copy['domain_tld'] = dns_df_a_copy['rrname'].apply(obtener_TLD)

# Eliminar todas las demás columnas excepto 'domain_tld'
dns_df_a_domain = dns_df_a_copy[['domain_tld']]

# Mostrar dataframe con la nueva columna
print("Dataframe con nueva columna 'domain_tld':")
dns_df_a_domain.head()
```

#### Parte 3 - Dominio experto

```python
# Convertir la columna 'isDGA' a tipo objeto
clasificacion_df['isDGA'] = clasificacion_df['isDGA'].astype(object)

# Concatenar los dataframes a lo largo de las columnas
dns_df_a_merged = pd.concat([dns_df_a, clasificacion_df], axis=1)

# Mostrar el dataframe combinado
print("Dataframe combinado:")
dns_df_a_merged
```

```python
# Cargar el archivo top-1m.csv en un DataFrame y convertir la columna "domain" a minúsculas
top_1m = pd.read_csv("top-1m.csv", names=["rank", "domain"])
top_1m["domain"] = top_1m["domain"].str.lower()

# Función para verificar si el TLD está en el top 1 millón de dominios
def is_in_top_1m(domain):
    return int(domain in top_1m["domain"].values)

# Crear una columna 'is_top_1m' en el DataFrame 'dga_dataframe'
def add_top_1m_column(dga_dataframe):
    dga_dataframe["is_top_1m"] = dga_dataframe["rrname"].apply(is_in_top_1m)
    return dga_dataframe

# Filtrar los dominios que están en el top 1 millón de dominios
def filter_top_1m(dga_dataframe):
    return dga_dataframe[dga_dataframe["is_top_1m"] == 0]["rrname"].unique()

# Filtrar los dominios DGA que no están en el top 1 millón oficial
def filter_dga_not_in_top_1m(dga_dataframe):
    return dga_dataframe[(dga_dataframe["isDGA"] == 0) & (dga_dataframe["is_top_1m"] == 1)]["rrname"].unique()

# Mostrar los dominios que parecen ser generados aleatoriamente (DGA)
def show_random_domains(dga_dataframe):
    dga_not_in_top_1m = filter_dga_not_in_top_1m(dga_dataframe)
    return [domain for domain in dga_not_in_top_1m if not domain.replace('.', '').isalnum()]
```

```python
# Agrega la columna is_top_1m al DataFrame dga_dataframe
dga_dataframe = add_top_1m_column(dns_df_a_merged)

# Filtra los dominios que están en el top 1 millón de dominios
top_1m_filtered = filter_top_1m(dga_dataframe)
print("Dominios en el top 1 millón:")
print(top_1m_filtered)

# Filtra los dominios DGA que no están en el top 1 millón oficial
dga_not_in_top_1m = filter_dga_not_in_top_1m(dga_dataframe)
print("\nDominios DGA que no están en el top 1 millón oficial:")
print(dga_not_in_top_1m)

# Muestra los dominios que parecen ser generados aleatoriamente (DGA)
random_domains = show_random_domains(dga_dataframe)
print("\nDominios que parecen ser generados aleatoriamente (DGA):")
print(random_domains)
```

#### Conclusion

En este laboratorio, hemos observado una diversidad de fuentes de dominios maliciosos, que van desde redes DGA (Domain Generation Algorithms) que generan dominios de forma aleatoria hasta dominios comprometidos utilizados para distribuir malware o phishing. Este amplio espectro de fuentes resalta la sofisticación y la variedad de las amenazas en línea a las que se enfrentan los sistemas de seguridad informática.

Es interesante notar que algunos de estos dominios maliciosos pueden encontrarse dentro del top 1 millón de dominios más buscados en Internet. Esta presencia en el top 1 millón puede deberse a tácticas de camuflaje por parte de los atacantes, quienes aprovechan la popularidad de ciertos dominios legítimos comprometidos para llevar a cabo sus actividades maliciosas sin levantar sospechas.

Un aspecto distintivo de los dominios generados por DGA es su naturaleza aleatoria. Esta característica los hace difíciles de predecir o identificar mediante patrones comunes, lo que los convierte en una herramienta atractiva para los atacantes en campañas de malware. La capacidad de generar dominios de manera no detectable facilita la comunicación con servidores de comando y control, lo que aumenta la dificultad para detectar y mitigar estas amenazas.

A través del análisis de datos y el filtrado de dominios utilizando herramientas como Pandas, podemos identificar patrones sospechosos y separar los dominios maliciosos de los legítimos. Este enfoque analítico es fundamental para la detección temprana y la mitigación de amenazas en entornos de ciberseguridad, ya que nos permite identificar y neutralizar dominios maliciosos antes de que puedan causar daño.
